```python
from nltk.corpus import brown

genre_word = [(genre, word)  # 生成 (类别, 单词) 二元组
              for genre in ["news", "romance"]  # 遍历两个类别
              for word in brown.words(categories=genre)]  # 遍历该类别下的所有单词

print(genre_word[:10])  # 查看前 10 个元素
```

```python
genre_word[-2:]
```

```python
from nltk import ConditionalFreqDist
cfd = nltk.ConditionalFreqDist(genre_word)
cfd
```

```python
cfd.conditions()
```

```python
cfd["romance"]["could"]
```

```python
cfd["news"]["could"]
```

```python
# 总统就职语料库
# how many conditions could we expect here?
# more than 2 (in this case they are actually 10),
# because we do not use the tokens in target as conditions, 
# but rather check for lower-cased words that start with these tokens
from nltk.corpus import inaugural
cfd = nltk.ConditionalFreqDist((w, fileid[:4])
             for fileid in inaugural.fileids()
              for w in inaugural.words(fileid)
               for target in ["american", "citizen"]
                if w.lower().startswith(target))
print(cfd.conditions())
```

```python
# Visualize cfd with: cfd.plot()
cfd.plot()
```

```python
# udhr – Universal Declaration of Human Rights Corpus: 
# the declaration of human rights in more than 300 languages
from nltk.corpus import udhr
languages = ["Afrikaans-Latin1", "Dutch_Nederlands-Latin1",
   "Danish_Dansk-Latin1", "Greenlandic_Inuktikut-Latin1",
   "Finnish_Suomi-Latin1", "Hungarian_Magyar-Latin1"]
cfd = nltk.ConditionalFreqDist(
   (lang, len(word))
   for lang in languages
   for word in udhr.words(lang)
)
cfd.plot()
```
